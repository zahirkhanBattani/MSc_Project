```python
import os
import pandas as pd
import numpy as np

print("Working dir:", os.getcwd())
print("Files in data/:")
print(os.listdir("../data") if os.path.exists("../data") else os.listdir("data"))
```

```python
# Cell 2 - load data
csv_path = "../data/project_risk_raw_dataset.csv"   # adjust filename as needed
df = pd.read_csv(csv_path)
df.head()
```

```python
# Cell 3 - quick EDA
df.info()          # column names, non-null counts, dtypes
df.describe(include='all').T   # stats
df.isna().sum().sort_values(ascending=False).head(20)  # missing values
```

```python
# Cell 4 - date parsing and simple features
for col in ['start_date', 'end_date', 'actual_end_date', 'created_at']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

# Example features (adjust names to your dataset)
if 'start_date' in df.columns and 'end_date' in df.columns:
    df['days_total'] = (df['end_date'] - df['start_date']).dt.days

# If there's a progress column
if 'progress' in df.columns:
    df['progress'] = pd.to_numeric(df['progress'], errors='coerce')

# Dependencies could be a string list; count items
if 'dependencies' in df.columns:
    df['num_dependencies'] = df['dependencies'].fillna('').apply(lambda s: len(s.split(',')) if s else 0)

# Safely describe only the columns that exist
cols_to_describe = [col for col in ['days_total', 'progress', 'num_dependencies'] if col in df.columns]
if cols_to_describe:
    display(df[cols_to_describe].describe())
else:
    print("No feature columns ('days_total', 'progress', 'num_dependencies') found to describe.")

```
